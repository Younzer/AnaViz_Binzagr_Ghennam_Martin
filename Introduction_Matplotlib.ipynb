### Exercice 2

```python
def sin_cos():
    x = np.linspace(-2*np.pi, 2*np.pi, 201)
    plot(x,np.sin(x),label='sin')
    plot(x,np.cos(x),label='cos')
    legend()
    
sin_cos()
```

```python
def plot_scatter(x,y):
    scatter(x,y)
    
plot_scatter(x,y)
```

```python
def plot_bar(x,height):
    bar(x,height)
    
b = range(5)
height = [1, 3, 3, 5, 4]
plot_bar(b,height)
```

```python
labels = np.array(['apples', 'foobar', 'cowboy', 'ananas', 'banana'])
plot_bar(labels,height)
```

```python
vals = np.random.randint(1000, size=(len(labels), 10))

def show_error_bar(vals,labels):
    means = np.mean(vals, axis=1)
    stds = np.std(vals, axis=1)
    bar(labels,means,yerr=stds)  
    
show_error_bar(vals,labels)
```

```python
def show_hist(array,n):
    hist(array,bins=n,density=1)
    
a = np.random.randint(100, size=1000)
show_hist(a,len(np.unique(a)))
```

```python
def moustache(vals,labels):
    boxplot(vals,labels=labels,vert=False) #,usermedians=np.std(vals,axis=0))
    
moustache(vals.T, labels)
```

```python
def bubble_chart(x,y,h):
    scatter(x,y,h)

h = np.random.randint(100, size=100)
bubble_chart(x,y,h)
```
